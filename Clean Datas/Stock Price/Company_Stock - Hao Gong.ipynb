```python
import datetime

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('company_stock.csv', index_col = 'Date', parse_dates=True)
#df
df1 = df.loc[:,['Close','Rate_of_Return','Volume','Company_name','Dis_category']]
df1
```

```python
###Analysis for all related companies### 

#close price 
test = pd.DataFrame(df1).reset_index()
df1_pivot = test.pivot(index = 'Date', columns = 'Company_name', values = 'Close')
df1_pivot = df1_pivot.dropna(how = 'all')
df1_pivot = df1_pivot['2010-01-01':'2017-12-24']
df1_pivot.head()
```

```python
#plt.plot(df1_pivot)
df1_pivot.plot(figsize=(16,14), title = 'Close Prices')
plt.show()
```

```python
# rate of return 

df2_pivot = test.pivot(index = 'Date', columns = 'Company_name', values = 'Rate_of_Return')
df2_pivot = df2_pivot.dropna(how = 'all')
df2_pivot = df2_pivot['2010-01-01':'2017-12-24']
df2_pivot.head()
```

```python
df2_pivot.plot(figsize=(16,14), title = 'Daily Rate of Return')
plt.ylim(-1, 2)
plt.show()
```

```python
###Disney stock analysis###

df_disney=df1.copy()
df_disney = df_disney.loc[df1['Dis_category']== 'Disney']#get disney data
#print(df_disney)

# Disney stock price with 50 and 200 moving average 
df_disney['MA50'] = df_disney['Close'].rolling(50).mean()
df_disney['MA200'] = df_disney['Close'].rolling(200).mean()
df_disney[['Close','MA50','MA200']].plot(figsize=(16,8),title = 'Disney Stock Price')
plt.show()

print('The Time of Highest Stock Price',':',df_disney['Close'].argmax())#the time of the highest price 
```

```python
#Disney stock trade volume 

df_disney['Volume'].plot(label='Disney',figsize=(16,8),title='Disney Volume Traded')
plt.show()

print('The Time of Highest Volume Trade',':',df_disney['Volume'].argmax())#the time of the highest trade volumn
```

```python
df_disney['Rate_of_Return'].plot.hist(bins = 100, title = 'Disney Rate of Return Histograms')
plt.show()
```

```python
###Selected fields : Studio Entertainment and Consumer_Products_&_Interactive_Media

hao_sub = df1.query("Dis_category != 'Media_Networks'").query("Dis_category != 'Parks_and_Resorts'")
hao_sub = hao_sub.reset_index()
hao_sub
#df_SE = df1.loc[df1['Dis_category'].isin(['Disney','Studio_Entertainment'])] 
#df_SE.head()
#df_SE = 

# ==== Close price Data set =====
hao_sub_close = hao_sub[['Date', 'Close', 'Company_name', 'Dis_category','Rate_of_Return']]
hao_sub_close_SE = hao_sub_close.query('Dis_category == "Studio_Entertainment"|Dis_category == "Disney"')
hao_sub_close_CP = hao_sub_close.query('Dis_category == "Consumer_Products_&_Interactive_Media"|Dis_category == "Disney"')


# ==== Rate of return data set ====
hao_sub_ret = hao_sub[['Date', 'Rate_of_Return', 'Company_name', 'Dis_category']]
hao_sub_ret_SE = hao_sub_ret.query('Dis_category == "Studio_Entertainment"|Dis_category == "Disney"')
hao_sub_ret_CP = hao_sub_ret.query('Dis_category == "Consumer_Products_&_Interactive_Media"|Dis_category == "Disney"')

print('Studio Entertainment',':',hao_sub_ret_SE.Company_name.unique())
print('Consumer_Products_&_Interactive_Media',':',hao_sub_ret_CP.Company_name.unique())
```

```python
# Select time range and pivot

hao_sub_close_SE_p = hao_sub_close_SE.pivot(index = 'Date', columns='Company_name', values = 'Close').loc['2010-01-01': '2017-10-24']
hao_sub_close_CP_p = hao_sub_close_CP.pivot(index = 'Date', columns='Company_name', values = 'Close').loc['2010-01-01': '2017-10-24']
hao_sub_ret_SE_p = hao_sub_ret_SE.pivot(index = 'Date', columns='Company_name', values = 'Rate_of_Return').loc['2010-01-01': '2017-10-24']
hao_sub_ret_CP_p = hao_sub_ret_CP.pivot(index = 'Date', columns='Company_name', values = 'Rate_of_Return').loc['2010-01-01': '2017-10-24']
```

```python
#Daily rate of return Boxplot_Studio_Entertainment - Risk Analysis
hao_sub_ret_SE_p.plot(kind = 'box', figsize = (16,11),
                           title = 'Boxplot for Studio Entertainment Companies')
plt.ylim(-0.1, 0.1)#get better view by ignoring extreme outliers 
plt.legend()
plt.show()
```

```python
#Daily rate of return_Consumer_Products_&_Interactive_Media
hao_sub_ret_CP_p.plot(figsize=(20,10), title = 'Daily Rate of Return - Consumer_Products_&_Interactive_Media')
plt.ylim(-0.3, 0.2)
l1 = plt.plot()
plt.show()
```

```python
#Daily rate of return Historgrams_Consumer_Products_&_Interactive_Media - Risk Analysis
hao_sub_ret_CP_p.plot.hist(bins = 300, figsize = (16,10),alpha = 0.7,
                           title = 'Rate of Return Histograms for Consumer_Products_&_Interactive_Media Companies')
plt.xlim(-0.2, 0.2)#get better view by ignoring extreme outliers 
plt.legend()
plt.show()
```

```python
#Daily rate of return KDE_Consumer_Products_&_Interactive_Media - Risk Analysis
hao_sub_ret_CP_p.plot(kind = 'kde', figsize = (16,10),alpha = 0.7,
                           title = 'KDE for Consumer_Products_&_Interactive_Media Companies')
plt.xlim(-0.2, 0.2)#get better view by ignoring extreme outliers 
plt.legend()
plt.show()
```

```python
#Daily rate of return Boxplot_Consumer_Products_&_Interactive_Media - Risk Analysis
hao_sub_ret_CP_p.plot(kind = 'box', figsize = (16,11),
                           title = 'Boxplot for Consumer_Products_&_Interactive_Media Companies')
plt.ylim(-0.1, 0.1)#get better view by ignoring extreme outliers 
plt.legend()
plt.show()
```

```python
#Std vs Return_Studio Entertainment

plt.scatter(hao_sub_ret_SE_p.mean(), hao_sub_ret_SE_p.std())

plt.xlim(-0.0005, 0.001)
plt.ylim(0.010, 0.030)
plt.xlabel('Daily Expected Return')
plt.ylabel('Daily Standard Deviation')
for label, x, y in zip(hao_sub_ret_SE_p.columns, hao_sub_ret_SE_p.mean(), hao_sub_ret_SE_p.std()):
    plt.annotate(
        label, xy = (x, y), xytext = (20,-20), textcoords = 'offset points', ha = 'right', 
        va = 'bottom', bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.6),
        arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
plt.show()
```
